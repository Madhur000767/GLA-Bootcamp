```python
import pandas as pd
df=pd.read_csv("session4_dummy_students_data.csv (1).xls")
```

```python
df["Average"] = df[["math", "science", "english"]].mean(axis=1)

print(df)
```

```python
df["Result"] = df["Average"].apply(lambda x: "Pass" if x >= 40 else "Fail")
print(df)
```

```python
class_avg = df["Average"].mean()
print("Class Average:", class_avg)
```

```python
print("Highest Average:", df["Average"].max())
print("Lowest Average:", df["Average"].min())
```

```python
top_students = df[df["Average"] >= 85]
print("\nTop Students:\n", top_students)
```

```python
low_attendance = df[df['attendance_percent'] < 75]

print("\nLow Attendance Students:\n", low_attendance)
```

```python
less_assignments = df[df['assignments_submitted'] < 6]
print("\nLess Assignments Submitted:\n", less_assignments)
```

```python
df.to_csv("session_student_report.csv", index=False)
```
