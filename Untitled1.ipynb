```python
import pandas as pd

# Read CSV file
df = pd.read_csv("session4_dummy_students.csv")

# Create average column using Math, Science, and English marks
df["Average"] = (df["Math"] + df["Science"] + df["English"]) / 3

# Create Result column
df["Result"] = df["Average"].apply(lambda x: "Pass" if x >= 40 else "Fail")

# Print class average
class_avg = df["Average"].mean()
print("Class Average:", round(class_avg, 2))

# Print highest and lowest average
print("Highest Average:", df["Average"].max())
print("Lowest Average:", df["Average"].min())

# Print top student(s)
top_avg = df["Average"].max()
top_students = df[df["Average"] == top_avg]

print("\nTop Student(s):")
print(top_students)

# Print students with attendance below 75
low_attendance = df[df["Attendance"] < 75]

print("\nStudents with Attendance below 75:")
print(low_attendance)

# Print students who submitted fewer than 6 assignments
few_assignments = df[df["Assignments_Submitted"] < 6]

print("\nStudents with fewer than 6 assignments:")
print(few_assignments)

# Save final output
df.to_csv("session4_student_report.csv", index=False)

print("\nFinal report saved as 'session4_student_report.csv'")
```
